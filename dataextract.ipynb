```python
import yfinance as yf
import pandas as pd
import numpy as np

# Date range
start = "2008-03-01"
end = pd.Timestamp.today().strftime("%Y-%m-%d")

# Yahoo Finance tickers
tickers = {
    "NIFTY": "^NSEI",
    "SP500": "^GSPC",
    "INDIAVIX": "^INDIAVIX",
    "CRUDE": "BZ=F",      # Brent Crude
    "USDINR": "INR=X"
}

dfs = {}

for name, ticker in tickers.items():
    print(f"Downloading {name}...")
    df = yf.download(
        ticker,
        start=start,
        end=end,
        auto_adjust=True,
        progress=False
    )

    close_df = df['Close']

    close_df.columns = [name]
    dfs[name] = close_df


data = pd.concat(dfs.values(), axis=1)


data = data.ffill()




data['LOG NIFTY'] = np.log(
    data['NIFTY'] / data['NIFTY'].shift(1)
)

data['LOG S&P500'] = np.log(
    data['SP500'] / data['SP500'].shift(1)
)

data['LOG CRUDE OIL'] = np.log(
    data['CRUDE'] / data['CRUDE'].shift(1)
)

data['LOG US DOLLAR'] = np.log(
    data['USDINR'] / data['USDINR'].shift(1)
)


data['INDIAVIX'] = data['INDIAVIX']


data = data.reset_index()
data['YEAR'] = data['Date'].dt.year
data['MONTH YEAR'] = data['Date'].dt.strftime('%m-%Y')

data.to_excel(
    "YahooFinance_Data.xlsx",
    index=False
)

print(data.head())
print("File saved successfully!")
```
